```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,30,40,50,60,70,75]
```

```python
weight_list = [20,60,70,93,96,102,75,63,84,94]
```

```python
plt.plot(age_list, weight_list, "r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
np_age_list = np.array(age_list)
np_weight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list, np_weight_list, "g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
numpy_arr1 = np.linspace(0,10,20)
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1, numpy_arr2, "b*-")
plt.show()
```

```python
plt.subplot(1,2,1) #iki grafiği yan yana koymaya yarıyor.
plt.plot(numpy_arr1, numpy_arr2, "b*-")
plt.subplot(1,2,2)
plt.plot(numpy_arr2, numpy_arr1, "r*-")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.3,0.3]) #grafiği büyütüp küçültmek.
figure_axes.plot(numpy_arr1, numpy_arr2, "g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
plt.show()
```

```python
new_fig = plt.figure(dpi = 100)
new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])

new_axes.plot(numpy_arr1, numpy_arr1 ** 2, label = "numpy array ** 2")
new_axes.plot(numpy_arr1, numpy_arr1 **3, label = "numpy array ** 3")
new_axes.legend(loc = 6) #legendin konumu
plt.show()
```

```python
data1 = np.linspace(0,10,20)
data2 = data1 **2
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1, data2, color = "#1E0703", alpha = 0.5)
my_axes.plot(data2, data1, color = "#5E2103")
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()

new_axes.plot(data1, data1 +1, color = "red", linewidth = 4.0)
new_axes.plot(data1, data1 +3, color = "yellow", linewidth = 10.0)

new_axes.plot(data1, data1 +4, color = "yellow", linestyle = "-.")
new_axes.plot(data1, data1 +4, color = "yellow", linestyle = ":", marker= "o", markersize = 8, markerfacecolor ="red")


plt.show()
```
