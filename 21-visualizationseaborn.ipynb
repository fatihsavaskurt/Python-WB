```python
plt.scatter("Height", "Weight", data = df_ev)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlete Weight vs Height")
plt.show()
```

```python
sns.set_style("white")
sns.scatterplot(x="Height", y="Weight", data = df_ev)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.show()
```

```python
sns.set_style("white")
sns.scatterplot(x="Height", y="Weight", hue="Sex", style = "Medal", data = df_ev) #hue kolona göre grafik çizdirir, medal kolona göre stil değiştirir.
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.show()
```
